```python
%matplotlib inline
```

```python
#import dependencies.
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare in the US dollars accumulated for each month. 
y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="d", color = "green", linewidth = 2, label='Boston')
#create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45.
plt.ylim(0, 45)
#create a title.
plt.title("PyBer Fare by Month")
#add a grid
plt.grid()
#add the legend. 
plt.legend()
```

```python
#create a bar chart
#create plot plot
plt.bar(x_axis, y_axis)
```

```python
#create the plot. 
plt.bar(x_axis, y_axis, color="green", label='Boston')
#create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create a title
plt.title("PyBer Fare by Month")
#add the legend. 
plt.legend()
```

```python
#create a horizontal bar chart
#create the plot
plt.barh(x_axis, y_axis)
```

```python
#switch the axes
#create the plot
plt.barh(y_axis, x_axis)
```

```python
#invert the axes
#create the plot
plt.barh(x_axis, y_axis, color = "magenta", label = 'Boston')
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
```

```python
#create a vertical bar chart
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#create a horizontal bar chart
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#Skill Drill
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label='Chicago')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#create a scotter plot

plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color = "red", label = 'Chicago')
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Creating Pie Charts
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#assign 12 colors, one for each month. 
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
